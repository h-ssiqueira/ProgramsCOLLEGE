```python
# Euclidiano
def calculoGuloso(atualX,atualY,possivelX,possivelY):
    return ((possivelX-atualX)**2 + (possivelY-atualY)**2)**0.5

# Manhattan
def calculoHeuristico(possivelX, possivelY, destinoX, destinoY):
    return abs(destinoX-possivelX) + abs(destinoY-possivelY)
```

```python
def buscaMenor(lista):
    #print(lista,"\n")
    menor = lista[0]
    for index in lista:
        if menor and menor[4] > index[4]:
            menor = index
    return menor
```

```python
# [posicao,closed,G,H,F,parent]
def AStarSearch(matriz, inicial, destino):
    matriz[inicial[0]][inicial[1]] = [inicial,1,0,0,0,None]
    naovisitados = []
    naovisitados.append(matriz[inicial[0]][inicial[1]])
    while naovisitados:
        menor = buscaMenor(naovisitados)
        naovisitados.remove(menor)
        matriz[menor[0][0]][menor[0][1]][1] = 1
        i = -1
        while i <= 1:
            j = -1
            while j <= 1:
                if menor[0][0]+i < len(matriz) and menor[0][0]+i >= 0 and menor[0][1]+j < len(matriz[0]) and menor[0][1]+j >= 0 and not(j == 0 and i == 0): # Verifica se posição existe
                    if matriz[menor[0][0]+i][menor[0][1]+j] == 0: # Verifica se já foi calculado o F na posição
                        if [menor[0][0]+i,menor[0][1]+j] == destino: # Verifica se chegou no destino
                            exibeCaminho(matriz,menor[0],destino)
                            return
                        matriz[menor[0][0]+i][menor[0][1]+j] = [
                            [menor[0][0]+i,menor[0][1]+j],
                            matriz[menor[0][0]+i][menor[0][1]+j],
                            calculoGuloso(menor[0][0],menor[0][1],menor[0][0]+i,menor[0][1]+j),
                            calculoHeuristico(menor[0][0]+i, menor[0][1]+j, destino[0], destino[1])
                        ]
                        matriz[menor[0][0]+i][menor[0][1]+j].append(matriz[menor[0][0]+i][menor[0][1]+j][3] + matriz[menor[0][0]+i][menor[0][1]+j][2])
                        matriz[menor[0][0]+i][menor[0][1]+j].append(menor[0])
                        naovisitados.append(matriz[menor[0][0]+i][menor[0][1]+j])
                    elif matriz[menor[0][0]+i][menor[0][1]+j] != 1 and matriz[menor[0][0]+i][menor[0][1]+j][4] > calculoGuloso(menor[0][0],menor[0][1],menor[0][0]+i,menor[0][1]+j) + calculoHeuristico((menor[0][0]+i), menor[0][1]+j, destino[0], destino[1]) and matriz[menor[0][0]+i][menor[0][1]+j][2] == 0: # Verifica se o F já calculado é menor que o F atual
                        matriz[menor[0][0]+i][menor[0][1]+j][1] = 0
                        matriz[menor[0][0]+i][menor[0][1]+j][2] = calculoGuloso(menor[0][0],menor[0][1],menor[0][0]+i,menor[0][1]+j)
                        matriz[menor[0][0]+i][menor[0][1]+j][3] = calculoHeuristico(menor[0][0]+i, menor[0][1]+j, destino[0], destino[1])
                        matriz[menor[0][0]+i][menor[0][1]+j][4] = matriz[menor[0][0]+i][menor[0][1]+j][3] + matriz[menor[0][0]+i][menor[0][1]+j][2]
                        matriz[menor[0][0]+i][menor[0][1]+j][5] = menor[0]
                        naovisitados.append(matriz[menor[0][0]+i][menor[0][1]+j])
                j = j + 1
            i = i + 1
```
